# Lambert Solver Verification (Hohmann Test Case)
This notebook verifies the Lambert solver against a direct Hohmann transfer from Caelus to Ventus.

```python
import numpy as np
from src.constants import *
from src.lambert import solve_lambert
from src.delta_v import departure_dv, arrival_capture_dv

# Analytical Hohmann transfer from Caelus to Ventus
r1 = 149.6e6  # km (assume approx 1 AU for Caelus)
r2 = 2.5 * r1   # km (assume approx 2.5 AU for Ventus)

# Time of flight is half period
a_transfer = (r1 + r2) / 2
tof_hohmann = np.pi * np.sqrt(a_transfer**3 / MU_STAR)

print(f'Hohmann TOF: {tof_hohmann/86400:.2f} days')

# Use Lambert Solver
r1_vec = np.array([r1, 0, 0])
r2_vec = np.array([-r2, 0, 0])
v1, v2 = solve_lambert(r1_vec, r2_vec, tof_hohmann, MU_STAR, direction='prograde')

print(f'Lambert V1 magnitude: {np.linalg.norm(v1):.3f} km/s')
print(f'Lambert V2 magnitude: {np.linalg.norm(v2):.3f} km/s')
```
